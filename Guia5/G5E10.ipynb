## a)

Considero $N$ mediciones $(x_i , y_i )$, donde los $y_i$ son independientes y todos con el mismo $\sigma$, y los $x_i$ son números sin error. Los parámetros de la recta $y=a_1+a_2 x$ que mejor ajusta los datos surgen de la fórmula de cuadrados mínimos:   
   

$\hat a_1 = (\sum x_i ^2 \sum y_i - \sum x_i \sum x_i y_i) / \Delta$   
$\hat a_2 = (N \sum x_i y_i - \sum x_i \sum y_i) / \Delta$   

$\textrm{con } \Delta = N \sum x_i ^2 - (\sum x_i )^2 $

   
Como los $x_i$ no son variable aleatorias, $a_1 = a_1(y_1, ..., y_n)$ y $a_2 = a_2(y_1, ..., y_n)$.    
Para calcular la matriz de covarianza entre estos parámetos, uso la fórmula de propagación de errores:   

$$Cov(\hat a_1, \hat a_2) = \sum_{i=1}^{N}\sum_{j=1}^{N} \frac{\partial \hat a_1}{\partial y_i} \frac{\partial \hat a_2}{\partial y_j} Cov(y_i, y_j) = \sum_{i=1}^{N} \frac{\partial \hat a_1}{\partial y_i} \frac{\partial \hat a_2}{\partial y_i} \sigma ^2 \qquad \textrm{pues} \qquad Cov(y_i, y_j) = \delta_{ij} \sigma ^2 $$

Calculo primero la varianza de $\hat a_1$:

$$ Var(\hat a_1) = Cov(\hat a_1, \hat a_1) = \sum_{k=1}^{N} (\frac{\partial \hat a_1}{\partial y_k}) ^2 \sigma ^2 $$


$\frac{\partial \hat a_1}{\partial y_k} = \frac{1}{\Delta} (\sum x_i - x_k \sum x_i) \qquad  \textrm{pues} \qquad \frac{\partial \sum y_i}{\partial y_k} = 1 \qquad \textrm{y} \qquad \frac{\partial \sum x_i y_i}{\partial y_k} = x_k$


$Var(\hat a_1) = \sum_{k=1}^{N} (\frac{\partial a_1}{\partial y_k}) ^2 \sigma ^2  = \frac{\sigma ^2}{\Delta^2} \sum_{k=1}^{N} (\sum x_i ^2 - x_k \sum x_i)^2  = \frac{\sigma ^2}{\Delta^2} \sum_{k=1}^{N} \big[(\sum x_i ^2)^2 -2 x_k \sum x_i ^2 \sum x_i + x_k ^2 (\sum x_i)^2 \big] \\ = \frac{\sigma ^2}{\Delta^2} \big[ N (\sum x_i ^2) ^2 - (\sum x_i) ^2 \sum x_i ^2 \big] = \frac{\sigma ^2}{\Delta^2} \big[ \sum x_i ^2 \big(N (\sum x_i ^2) - (\sum x_i) ^2 \big) \big] = \frac{\sigma ^2}{\Delta^2} \Delta \sum x_i ^2  = \frac{\sigma ^2}{\Delta} \sum x_i ^2 $
  
   
Calculo ahora la varianza de $\hat a_2$:

$$ Var(\hat a_2) = \sum_{k=1}^{N} (\frac{\partial \hat a_2}{\partial y_k}) ^2 \sigma ^2 $$

$\frac{\partial \hat a_2}{\partial y_k} = \frac{1}{\Delta} \big[ N \frac{\partial }{\partial y_k} ( \sum x_i y_i) - \sum x_i  \frac{\partial }{\partial y_k} (\sum y_i) \big] = \frac{1}{\Delta} (N x_k - \sum x_i) $

$Var(\hat a_2) = \sum_{k=1}^{N} \frac{\partial \hat a_2}{\partial y_k}) ^2 \sigma ^2 = \sum_{k=1}^{N} \frac{\sigma^2}{\Delta^2} (N x_k - \sum x_i) ^2  = \frac{\sigma^2}{\Delta^2} \sum_{k=1}^{N} \big[ N^2 x_k ^2 -2N x_k \sum x_i + (\sum x_i)^2  \big] \\ = \frac{\sigma^2}{\Delta^2} \big[ N^2 \sum x_i ^2 -2N(\sum x_i)^2 +N(\sum x_i)^2  \big] = \frac{\sigma^2}{\Delta^2} N \Delta = \frac{\sigma^2}{\Delta} N$
  
   
Por último, calculo la covarianza: 

$\frac{\partial \hat a_2}{\partial y_i} = \frac{1}{\Delta} (\sum x_i - x_k \sum x_i)\frac{1}{\Delta} (N x_k - \sum x_i)  = \frac{1}{\Delta^2} \big( N x_k \sum x_i ^2 - \sum x_i ^2\sum x_i - N x_k ^2 \sum x_i +x_k (\sum x_i) ^2 \big)$

$ Cov(\hat a_1, \hat a_2) = \sum_{i=1}^{N} \frac{\partial \hat a_1}{\partial y_i} \frac{\partial \hat a_2}{\partial y_i} \sigma ^2 = \frac{\sigma^2}{\Delta^2} \sum_{k=1}^{N} \big( N x_k \sum x_i ^2 - \sum x_i ^2\sum x_i - N x_k ^2 \sum x_i +x_k (\sum x_i) ^2 \big) \\ = \frac{\sigma^2}{\Delta^2} \big( N \sum x_i \sum x_i ^2 - N \sum x_i \sum x_i ^2 -  N \sum x_i \sum x_i ^2 +\sum x_i ( \sum x_i)^2  \big) \\ = - \frac{\sigma^2}{\Delta^2} \sum x_i \big( \sum x_i ^2 - (\sum x_i) ^2  \big)  = - \frac{\sigma^2}{\Delta^2} \sum x_i \Delta  = - \frac{\sigma^2}{\Delta} \sum x_i$

## b)

Veamos qué pasa con la correlación entre la pendiente y la ordenada al origen si el promedio de los datos está del lado positivo de la abscisa. Para eso uso los datos del ítem siguiente:

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython import display
import random as rd
%matplotlib inline
```

```python
x = np.linspace(1.90,3.10,100) #para graicar rectas
x_dat = [2.00, 2.10, 2.20, 2.30, 2.40, 2.50, 2.60, 2.70, 2.80, 2.90, 3.00]
y_dat = [2.78, 3.29, 3.29, 3.33, 3.23, 3.69, 3.46, 3.87, 3.62, 3.40, 3.99]
sigmay = 0.3 #el error del promedio es sigma/(raiz de N)
error, errorprom = sigmay*np.ones(len(y_dat)), sigmay/(np.sqrt(len(y_dat)))
x_prom, y_prom = round(np.mean(x_dat),2), round(np.mean(y_dat),2)
x_min, x_max = np.min(x_dat), np.max(x_dat)
y_min, y_max = np.min(y_dat), np.max(y_dat)
```

```python
fig = plt.figure(figsize=(20,5))
plt.plot(x_dat, y_dat, 'go', label= 'Datos')
plt.errorbar(x_dat, y_dat, error, ecolor='k', linestyle = 'None')
plt.plot(x_prom, y_prom, 'bs', label = 'Promedio')
plt.errorbar(x_prom, y_prom, errorprom, ecolor='c', linestyle = 'None')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlim([x_min-0.1,x_max+0.1])
plt.ylim([y_min-sigmay,y_max+sigmay])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Grafico con promedio positivo')
plt.grid()

pendiente = -1
while(pendiente<=1.40): #vario la pendiente
    ordenada = y_prom - pendiente*x_prom
    plt.plot(x, pendiente*x+ordenada,'m-', alpha = 0.7)
    display.clear_output(wait=True)
    display.display(plt.gcf())
    plt.draw()
    pendiente=pendiente+0.05
```

Como se observa en el gráfico, si aumenta la ordenada al origen disminuye la pendiente. Esto significa que la correlación entre ambas es negativa.   
   
Veamos qué pasa si el promedio está del lado negativo de la abscisa:

```python
x_dat_neg =  [ -d for d in x_dat]
x_neg = np.linspace(-3.10,-1.90,100)
x_prom_neg= round(np.mean(x_dat_neg),2)
x_min_neg, x_max_neg = np.min(x_dat_neg), np.max(x_dat_neg)

fig = plt.figure(figsize=(20,5))
plt.plot(x_dat_neg, y_dat, 'go', label= 'Datos')
plt.errorbar(x_dat_neg, y_dat, error, ecolor='k', linestyle = 'None')
plt.plot(x_prom_neg, y_prom, 'bs', label = 'Promedio')
plt.errorbar(x_prom_neg, y_prom, errorprom, ecolor='c', linestyle = 'None')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlim([x_min_neg-0.1,x_max_neg+0.1])
plt.ylim([y_min-sigmay,y_max+sigmay])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Grafico con promedio negativo')
plt.grid()

pendiente = -1
while(pendiente<=1.40): #vario la pendiente
    ordenada = y_prom - pendiente*x_prom_neg
    plt.plot(x_neg, pendiente*x_neg+ordenada,'m-', alpha = 0.7)
    display.clear_output(wait=True)
    display.display(plt.gcf())
    plt.draw()
    pendiente=pendiente+0.05
```

A pesar del aparente parecido con el gráfico anterior, se debe notar que en éste se obervan los valores negativos de x, entonces del lado derecho va aumentando la ordenada al origen a medida que, simultáneamente, también aumenta la pendiente. Esto significa que los parámetros tienen una correlación positiva.   
   
Por último, veamos qué pasa si el promedio es cero:

```python
x_cero = np.linspace(-0.6,0.6,100)
x_dat_cero =  [ d-(x_prom) for d in x_dat]
x_prom_cero= round(np.mean(x_dat_cero),2) #o sea, cero
x_min_cero, x_max_cero = np.min(x_dat_cero), np.max(x_dat_cero)

fig = plt.figure(figsize=(20,5))
plt.plot(x_dat_cero, y_dat, 'go', label= 'Datos')
plt.errorbar(x_dat_cero, y_dat, error, ecolor='k', linestyle = 'None')
plt.plot(x_prom_cero, y_prom, 'bs', label = 'Promedio')
plt.errorbar(x_prom_cero, y_prom, errorprom, ecolor='c', linestyle = 'None')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlim([x_min_cero-0.1,x_max_cero+0.1])
plt.ylim([y_min-sigmay,y_max+sigmay])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Grafico con promedio cero')
plt.grid()

pendiente = -1
while(pendiente<=1.40): #vario la pendiente
    ordenada = y_prom - pendiente*x_prom_cero
    plt.plot(x_cero, pendiente*x_cero+ordenada,'m-', alpha = 0.7)
    display.clear_output(wait=True)
    display.display(plt.gcf())
    plt.draw()
    pendiente=pendiente+0.05
```

En este caso, aunque varía la pendiente, la ordenada al orgien permanece constante, es decir, no hay correlación entre ambos parámetros.

## c)

Para ajustar los datos utilizando el método de cuadrados mínimos, defino una función que calcula los parámetros $\hat a_1$ y $\hat a_2$, para datos donde solamente los $y$ tienen error (es decir, los $x$ no son variables aleatorias).

```python
def cuadradosmin(x,y): #solo con error en y
    delta = len(y)*np.sum(x**2) - np.sum(x)**2
    a1 = (np.sum(x**2)*np.sum(y)-np.sum(x)*np.sum(x*y))/delta
    a2 = (len(y)*np.sum(x*y) - np.sum(x)*np.sum(y))/delta
    return [a1, a2, delta]
```

```python
x_array, y_array = np.array(x_dat), np.array(y_dat) #los convierto en array para poder usarlos en la funcion
parametros = cuadradosmin(x_array,y_array)
a1, a2 = round(parametros[0],3), round(parametros[1],3)
var_a1 = ((sigmay**2)/parametros[2])*np.sum(x_array**2)
var_a2 = ((sigmay**2)/parametros[2])*len(y_array)
cov_a1a2 = -((sigmay**2)/parametros[2])*np.sum(x_array)
sigma_a1, sigma_a2 = round(np.sqrt(var_a1),3), round(np.sqrt(var_a2),3)

print('a1 = {} +- {} \na2 = {} +- {}').format(a1, sigma_a1, a2, sigma_a2)
```

Grafico ahora la recta obtenida con estos parámetros:

```python
xc = np.linspace(0,5,100)
fig = plt.figure(figsize=(20,5))
plt.plot(x_dat, y_dat, 'go', label= 'Datos')
plt.errorbar(x_dat, y_dat, error, ecolor='k', linestyle = 'None')
plt.plot(xc, parametros[1]*xc+parametros[0], 'b-', label = 'Ajuste lineal')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlim([0,5])
plt.ylim([0,6])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ajuste lineal de los datos utilizando el metodo de cuadrados minimos')
plt.grid()
```

## d)

Ahora, una vez conocidos los parámetros $a_1$ y $a_2$, puedo definir una recta $y_a=a_1+a_2 x_a$, donde las variables aleatorias son $a_1$ y $a_2$ (y por lo tanto, $y_a$), pero no $x_a$. Para un $x_a$ dado, puedo calcular su $y_a$ correspondiente, cuya varianza viene dada por

$Var(y_a) = \big( \frac{\partial y_a}{\partial a_1} \big) ^2 Var(a_1) + \big( \frac{\partial y_a}{\partial a_2} \big) ^2 Var(a_2) + 2 \frac{\partial y_a}{\partial a_1} \frac{\partial y_a}{\partial a_2} Cov(a_1,a_2) \\ = Var(a_1) + x_a^2 Var(a_2) +2 x_a Cov(a_1, a_2) = \frac{\sigma^2}{\Delta}(\sum x_i ^2 + x_a ^2 N -2 x_a \sum x_i) \equiv \sigma_{y_a}$

   
Tomo un valor arbitrario para $x_a$ y lo agrego al gráfico:

```python
xa_d = rd.uniform(0,5) #tomo un x_a cualquiera
ya_d  = parametros[0] + parametros[1]*xa_d #calculo su y_a
sigmaya_d = np.sqrt(var_a1 + (xa_d**2)*var_a2 + 2*xa_d*cov_a1a2) #con su error
fig = plt.figure(figsize=(20,5))
plt.plot(x_dat, y_dat, 'go', label= 'Datos')
plt.errorbar(x_dat, y_dat, error, ecolor='k', linestyle = 'None')
plt.plot(xc, parametros[1]*xc+parametros[0], 'b-', label = 'Ajuste lineal')
plt.plot(xa_d, ya_d, 'ro', label ='Punto agregado')
plt.errorbar(xa_d, ya_d, sigmaya_d, ecolor = 'm', linestyle='None')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlim([0,5])
plt.ylim([0,6])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ajuste lineal de los datos con un punto agregado')
plt.grid()
```

Al repetir la celda anterior, se observa claramente que no todos los puntos agregados tienen el mismo error. Busco qué $x_a$ minimiza el error de $y_a$:

$0= \frac{\partial \sigma_{y_a}^2}{\partial x_a} = \frac{\sigma^2}{\Delta}(2 x_a N - 2 \sum x_i) \qquad \Rightarrow \qquad x_a = \frac{\sum x_i}{N} $

El $x_a$ que corresponde al $y_a$ de menor error es el promedio.   
   
Agrego ahora puntos sobre toda la recta, con sus corresponidentes errores, para ver cómo varían a medida que se alejan de la región dodne se hicieron las mediciones.

```python
xa_agregados = np.linspace(0,5,70)
ya_agregados = parametros[0] + parametros[1]*xa_agregados
sigmaya_agregados = np.sqrt(var_a1 + (xa_agregados**2)*var_a2 + 2*xa_agregados*cov_a1a2)

fig = plt.figure(figsize=(20,5))
plt.plot(x_dat, y_dat, 'go', label= 'Datos')
plt.errorbar(x_dat, y_dat, error, ecolor='k', linestyle = 'None')
plt.plot(xc, parametros[1]*xc+parametros[0], 'b-', label = 'Ajuste lineal')
plt.plot(xa_agregados, ya_agregados, 'ro', label ='Puntos agregados')
plt.errorbar(xa_agregados, ya_agregados, sigmaya_agregados, ecolor = 'm', linestyle='None')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlim([0,5])
plt.ylim([0,6])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ajuste lineal de los datos con puntos agregados')
plt.grid()
```

Como era de esperar, los puntos agregados tienen menor error cuanto más cerca se encuentran del promedio de los datos, y aumentan a medida que se alejan de los datos. Este resultado tiene sentido, dado que la región que "mejor conocemos" (es decir, que podemos predecir con mayor precisión) es donde están los datos, y "conocemos cada vez menos" cuanto más nos alejamos de ellos.

## e)

Veamos qué pasa si no tenemos en cuenta la covarianza a la hora de calcular el error de $y_a$:

```python
sigmaya_sincov = np.sqrt(var_a1 + (xa_agregados**2)*var_a2)

fig = plt.figure(figsize=(20,5))
plt.plot(x_dat, y_dat, 'go', label= 'Datos')
plt.errorbar(x_dat, y_dat, error, ecolor='k', linestyle = 'None')
plt.plot(xc, parametros[1]*xc+parametros[0], 'b-', label = 'Ajuste lineal')
plt.plot(xa_agregados, ya_agregados, 'ro', label ='Puntos agregados')
plt.errorbar(xa_agregados, ya_agregados, sigmaya_sincov, ecolor = 'm', linestyle='None')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlim([0,5])
plt.ylim([0,7])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ajuste lineal de los datos con puntos agregados')
plt.grid()
```

El error de $y_a$ aumenta a medida que aumenta $x_a$, sin tener en cuenta dónde están los datos. Esto ocurre porque la ecuación para el $\sigma_{y_a}^2$ es cuadrática, cuyo mínimo, teniendo en cuenta la covarianza, se encuentra en el promedio de los datos, mientras que sin tener en cuenta la covarianza,se encuentra en el cero (recordar que la covarianza es negativa):


$\sigma_{y_a}^2 = Var(a_1) + x_a^2 Var(a_2) +2 x_a Cov(a_1, a_2) \qquad \textrm{con covarianza}$

$\sigma_{y_a}^2 = Var(a_1) + x_a^2 Var(a_2)  \qquad \textrm{sin covarianza}$
 

```python
sigma2_concov, sigma2_sincov = sigmaya_agregados**2, sigmaya_sincov**2
sigma2_concov_prom, sigma2_sincov_prom = var_a1 + (x_prom**2)*var_a2 + 2*x_prom*cov_a1a2, var_a1 + (x_prom**2)*var_a2

fig = plt.figure(figsize=(15,6))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)
fig.subplots_adjust(hspace=0.6)

ax1.plot(xa_agregados, sigma2_concov, 'b-')
ax1.plot(x_prom, sigma2_concov_prom, 'go')
ax1.set_xlabel('xa')
ax1.set_ylabel('Var(ya)')
ax1.set_title('Con covarianza')
ax1.grid()

ax2.plot(xa_agregados, sigma2_sincov, 'b-')
ax2.plot(x_prom, sigma2_sincov_prom, 'go', label = 'xa promedio')
ax2.set_xlabel('xa')
ax2.set_ylabel('Var(ya)')
ax2.set_title('Sin covarianza')
ax2.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
ax2.grid()

```

El segundo gráfico no tiene sentido: El error es más chico en el cero, donde no hay ningún dato, que donde sí hay, esto significaría que conocemos con más precisión lo que esperamos obtener en una región en la cual no tenemos datos. En cambio, en el primer gráfico, se observa claramente que el mínimo error se encuentra en el $x$ promedio de los datos.
